```python
import pandas as pd
import plotly.graph_objects as go

from config import filename_excel
```

```python
df = pd.read_excel(filename_excel)
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%d/%m/%Y')
```

```python
df['Day_Name'] = df['Date'].dt.day_name()
```

```python
filt = df['Date'] >= df['Date'].max() - pd.DateOffset(years=1)
df_year = df[filt]
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
                x = df_year['Date'],
                y = df_year['first'],
                name = "First Winning Number",
                line_color = "deepskyblue",
                mode = "lines+markers",
                opacity = 0.8))

fig.add_trace(go.Scatter(
                x = df_year['Date'],
                y = df_year['second'],
                name = "Second Winning Number",
                line_color = "yellow",
                mode = "lines+markers",
                opacity = 0.8))

fig.add_trace(go.Scatter(
                x = df_year['Date'],
                y = df_year['third'],
                name = "Third Winning Number",
                line_color = "red",
                mode = "lines+markers",
                opacity = 0.8))

fig.add_trace(go.Scatter(
                x = df_year['Date'],
                y = df_year['fourth'],
                name = "Fourth Winning Number",
                line_color = "green",
                mode = "lines+markers",
                opacity = 0.8))

fig.add_trace(go.Scatter(
                x = df_year['Date'],
                y = df_year['fifth'],
                name = "Fifth Winning Number",
                line_color = "violet",
                mode = "lines+markers",
                opacity = 0.8))

fig.add_trace(go.Scatter(
                x = df_year['Date'],
                y = df_year['sixth'],
                name = "Sixth Winning Number",
                line_color = "black",
                mode = "lines+markers",
                opacity = 0.8))

fig.update_layout(xaxis_range=[df_year['Date'].min(), df_year['Date'].max()])

fig.update_xaxes(
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = "1 month", step = "month", stepmode = "backward"),
            dict(count = 6, label = "6 months", step = "month", stepmode = "backward"),
            dict(step="all")
        ])
    )            
)

fig.show()
```
